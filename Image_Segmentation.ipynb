```python
from sklearn.cluster import KMeans
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
%matplotlib inline
```

```python
img = Image.open("/content/drive/MyDrive/Data Science/Lesson 7 Unsupervised/Lesson 7-Unsupervised Learning/tiger.png")
img_np = np.array(img)
img_np[0:2]
```

```python
img_np.shape
```

```python
pixel = img_np.reshape(img_np.shape[0]*img_np.shape[1],img_np.shape[2])
pixel.shape
```

```python
kmeans = KMeans(n_clusters=16)
kmeans.fit(pixel)
```

After the model is trained, model.labels_ is used to obtain the number of <br>cluster that is assigned to each data point or each pixel.<br>
model.cluster_centers_ gives us the coordinates or the RGB values of the 16 cluster centers.

```python
pixel_centroids = kmeans.labels_
cluster_pix_centers  = kmeans.cluster_centers_
pixel_centroids
```

```python
cluster_pix_centers
```

```python
final=np.zeros((pixel_centroids.shape[0],3))
for cluster_no in range(16):
  final[pixel_centroids==cluster_no]=cluster_pix_centers[cluster_no]
  print(final.shape)
final.shape
```

```python
final.shape
```

```python
pixel_centroids.shape
```

```python
comp_image=final.reshape(img_np.shape[0],img_np.shape[1],3)
comp_image.shape
```

```python
comp_image=Image.fromarray(np.uint8(comp_image))
comp_image.save('/content/drive/MyDrive/Data Science/Lesson 7 Unsupervised/Lesson 7-Unsupervised Learning/tiger_compressed.png')
img_1 = mpimg.imread('/content/drive/MyDrive/Data Science/Lesson 7 Unsupervised/Lesson 7-Unsupervised Learning/tiger.png')
img_2 = mpimg.imread('/content/drive/MyDrive/Data Science/Lesson 7 Unsupervised/Lesson 7-Unsupervised Learning/tiger_compressed.png')
```
