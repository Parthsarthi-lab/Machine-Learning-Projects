<b> Applying label encoder on the categorical variables

```python
# Label encoder on train set
for column in df_train.select_dtypes(include=["object"]).columns:
    le = LabelEncoder()
    df_train[column] = le.fit_transform(df_train[column])
print(df_train.select_dtypes(include=["object"]).dtypes)  # Shows that no object data type remains
```

```python
# Label encoder on test set
for column in df_test.select_dtypes(include=["object"]).columns:
    le = LabelEncoder()
    df_test[column] = le.fit_transform(df_test[column])
print(df_test.select_dtypes(include=["object"]).dtypes)  # Shows that no object data type remains
```

<b> Performing Dimesionality Reduction

Using Principal Component Analysis

```python
# Train Test Split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(df_train.drop(['y'],axis=1),df_train.y,test_size=0.3,random_state=1)

print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
# Selecting the continuous variables
X_train_num =  xtrain.drop(labels=["ID","X0","X1","X2","X3","X4","X5","X6","X8"],axis=1)
X_test_num = xtest.drop(labels=["ID","X0","X1","X2","X3","X4","X5","X6","X8"],axis=1)
```

```python
print(X_train_num.shape)
print(X_test_num.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train_num)
scaled_train = scaler.transform(X_train_num)
scaled_test = scaler.transform(X_test_num)
```

```python
from sklearn.decomposition import PCA
pca= PCA(n_components=0.95)
pca.fit(scaled_train)
```

```python
x_train_trans = pca.transform(scaled_train)
print("Orignal Shape :  ",xtrain.shape)
print("Transformed Shape :  ",x_train_trans.shape)
```

```python
x_test_trans = pca.transform(scaled_test)
print("Orignal Shape :  ",xtest.shape)
print("Transformed Shape :  ",x_test_trans.shape)
```

```python
xtrain
```

```python
# Changing xtrain -> ndarray -> dataframe # IMPORTANT
xtrain = pd.concat([pd.DataFrame(x_train_trans),pd.DataFrame(np.array(xtrain[["ID","X0","X1","X2","X3","X4","X5","X6","X8"]]),columns=["ID","X0","X1","X2","X3","X4","X5","X6","X8"])],axis=1)
xtrain
```

```python
# Changing xtrain -> ndarray -> dataframe # IMPORTANT
xtest = pd.concat([pd.DataFrame(x_test_trans),pd.DataFrame(np.array(xtest[["ID","X0","X1","X2","X3","X4","X5","X6","X8"]]),columns=["ID","X0","X1","X2","X3","X4","X5","X6","X8"])],axis=1)
xtest
```
